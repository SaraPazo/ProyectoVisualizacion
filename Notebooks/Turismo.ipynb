```python
import pandas as pd
import numpy as np
import statistics as stats

import warnings
warnings.filterwarnings('ignore')

import pylab as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Data/Draft_data/viajes.csv', encoding='latin1', sep=';')
data.head()
```

```python
len(data)
```

Este trata sobre el turismo en la Comunidad de Madrid, por trimestre, seleccionando el sexo y la edad. 

-  Elimino todas aquellas filas que tengan nulos (NaN) en la columna 'Total', ya que no me servirán para el estudio. Estos nulos indican que en estos últimos trimestres no se han recopilado todavía datos.

```python
data = data.dropna(subset=['Total'])

# Mostrar el DataFrame después de eliminar las filas con NaN en 'Total'
data.head()
```

- Elimino la columna 'Tipo de dato' ya que es un valor que lleva consigo la tabla por su selección previa. Sin embargo, no nos será util. 

```python
data = data.drop('Tipo de dato', axis=1)

data.head()
```

Elimino las filas de la columna 'Sexo' siembre que el dato de la fila sea 'Total'. 
Lo que me interesa es saber si es hombre o mujer, ya que el total será la suma de ambos, en cada trimestre. 

```python
data = data[data['Sexo'] != 'Total']

data.head()
```

```python
import re
```

```python
data['Edad'] = data['Edad'].astype(str)

# Extraer los valores numéricos de la columna 'Edad' usando expresiones regulares
data['Edad'] = data['Edad'].apply(lambda x: re.findall(r'\d+', x))

# Crear una nueva columna 'Rango de Edad' utilizando el rango extraído
data['edad'] = data['Edad'].apply(lambda x: f"{x[0]}-{x[1]}" if len(x) == 2 else None)
```

```python
data
```

```python
data = data.drop('edad', axis=1)
```

```python
data[['Año', 'Trimestre']] = data['Periodo'].str.split('T', expand=True)

data.head()
```

```python
turismo = data.rename(columns={
    'Sexo': 'Genero',
    'Edad': 'Rango_Edad',
    'Concepto turÃ­stico': 'Concepto_turistico',
    'Total': 'Total_Visitas'
})
```

```python
turismo = turismo.reset_index(drop=True)

turismo.head()
```

```python
nuevo_orden = ['Concepto_turistico', 'Genero', 'Rango_Edad', 'Año', 'Trimestre', 'Periodo', 'Total_Visitas']

# Reorganizar el DataFrame con el nuevo orden de columnas
turismo = turismo[nuevo_orden]
```

```python
turismo.insert(0, 'comunidad', 'Madrid')
turismo.head()
```
